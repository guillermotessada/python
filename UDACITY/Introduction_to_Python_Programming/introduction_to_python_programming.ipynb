# String Methods
* los metodos estan relacionados con las funciones
* *los metodos son asocioados a especificos objetos
* hay diferentes metodos dependiendo del del tipo de objeto
* visto que los metodos son especificos de un objeto siempre estara delanta el objeto seguido por un metodo: objeto.metodo( )
* hay metodos que necesitan un argumento como count(argumento)

  

```python
# .title() es un metodo para un string, lo que hace es 
# poner mayuscula despues de un espacio
print('guillermo tessada'.title())
print('guillermotessada'.title())
```

```python
# .islower()  es un metodo que verifica si todas las palabras
# estan en minusculas y retorna un bool si todas son minusculas es True
# si hay mayusculas False
print('Guillermotessada'.islower())
print('guillermotessasda'.islower())
```

```python
# .count() este metodo permite de contar occurrencias
# en este caso cuenta el numero de a en el cadena
x= 'dkfahdjfalflkhasfafissh'
print(x.count('a'))
print(x.count('ha'))
```

# Metodo split()
* Un método de cadena útil cuando se trabaja con cadenas es el método .split. 
Esta función o método devuelve un contenedor de datos <br> llamado lista que contiene las palabras de la cadena de entrada
* El método split tiene dos argumentos adicionales (sep y maxsplit). El argumento sep significa "separador". <br>Se puede utilizar para identificar cómo se debe dividir la cadena (por ejemplo, caracteres de espacio en blanco como espacio, tabulación,

```python
# el metodo .split() entrega una lista con las cadenas del string
new_str = "The cow jumped over the moon."
new_str.split()
```

```python
#metodo split indicando el separador y el hasta donde llega el split
new_str = "The cow jumped over the moon."
print(new_str.split(' ',3)) # hace el split hasta el 3er espacio
print(new_str.split(' ', 1)) # hace el split hasta el primer espacio
```

```python
new_str = "The cow .jumped over. the moon"
new_str.split('.') # toma como referencia para el split el '.'
```
