# If, Elif, Else - controlando el flujo

* **if**: es una declaracin **if** siempre debe comenzar con una clusula **if**, que contiene la primera condicin que se verifica. Si esto se evala como Verdadero, Python ejecuta el cdigo sangrado en este bloque **if** y luego salta al resto del cdigo despus de la declaracin **if**

* **elif**: **elif** es la abreviatura de **"else if"**. Una clusula elif se utiliza para verificar una condicin adicional si las condiciones de las clusulas anteriores en la declaracin **if** se evalan como Falsas.

* **else**: la clusula **else**, que debe aparecer al final de una declaracin **if** si se usa. Esta clusula no requiere ninguna condicin. El cdigo en un bloque **else** se ejecuta si todas las condiciones anteriores en la declaracin **if** se evalan como False.

* **NOTA**: cuando se usa el control de flujo el codigo se lee desde el primer **if** y se para hasta no encontrar una condicion verdadera







```python
# uso de if y else
# numero%= 0 se usa para verificar si el numero es par o no
# en este caso la condicion en el primer if es False por lo tanto pasa a else
number = 145
if number % 2 == 0:
    print("Number " + str(number) + " es entero.")
else:
    print("Number " + str(number) + " no es entero.")
```

```python
# uso de if y else
# numero%= 0 se usa para verificar si el numero es par o no
# en este caso la condicion en el primer if es True por lo tanto para en el if
number = 300
if number % 2 == 0:
    print("Number " + str(number) + " es entero.")
else:
    print("Number " + str(number) + " no es entero.")
```

# For Loops
* permiten de repetir bloques de codigo
* Python tiene dos tipos de bucles: bucles **for** y bucles **while**. Un bucle **for** se utiliza para "iterar" o hacer algo repetidamente sobre un iterable.
* Un iterable es un objeto que puede devolver uno de sus elementos a la vez. Esto puede incluir tipos de secuencia, como cadenas, listas y tuplas, as como tipos que no son de secuencia, como diccionarios y archivos.
* estructura: 
    * for indica el ciclo
    * i es la variable de iteracion
    * list es el objeto iterable
* explicacion de los componentes de un ciclo for:
    * La primera lnea del bucle comienza con la palabra clave **for**, que indica que se trata de un bucle for.A continuacin city en cities, lo que indica que city es la variable de iteracin y cities es el iterable sobre el que se realiza el bucle. 
    * La lnea de encabezado del bucle for siempre termina con dos puntos **:**
    * En la primera iteracin del ciclo, city obtiene el valor del primer elemento de las ciites, que es "nueva york". 
    * Despus del encabezado del bucle for hay un bloque de cdigo con sangra, el cuerpo del bucle, que se ejecutar en cada iteracin de este bucle. 
    * Una vez que se ha ejecutado el cuerpo del bucle, todava no pasamos a la siguiente lnea; volvemos a la lnea de encabezado **for**, donde la variable de iteracin toma el valor del siguiente elemento del iterable. En la segunda iteracin del bucle anterior, la ciudad toma el valor del siguiente elemento en las ciudades, que es "mountain view".
    * Este proceso se repite hasta que el bucle haya recorrido todos los elementos del iterable. 
    * Luego, pasamos a la lnea que sigue al cuerpo del bucle; en este caso, print("Done!"). 
    * la ultima lnea despus del cuerpo del bucle no presente sangra.

```python
# ejemplo for en una lista
cities = ['new york city', 'mountain view', 'chicago', 'los angeles']
for city in cities:
    print(city)
print("Done!")
```

# Using the range() Function with for Loops
* **range(start, stop, step)** es una funcin incorporada que se utiliza para crear una secuencia iterable de nmeros. 
* se usa  **range()** con un bucle for para repetir una accin un determinado nmero de veces. 
* se puede usar cualquier variable para iterar a travs de los nmeros, pero los programadores de Python convencionalmente usan **i**
* en la funcion **range** **start** parte de 0 , mientras **stop** es incluido
* sintaxis:
    * La funcin range() toma tres argumentos, el primero y el tercero son opcionales:
    * El argumento 'start' es el primer nmero de la secuencia. Si no se especifica, "start" por defecto es 0.
    * El argumento 'stop' es 1 ms que el ltimo nmero de la secuencia. Este argumento debe especificarse.
    * El argumento del 'step' es la diferencia entre cada nmero de la secuencia. Si no se especifica, el valor predeterminado de "paso" es 1.
   <br> 
* Notas: 
    * Si especifica un nmero entero dentro del parntesis con rango(), se usa como valor para 'detener' y los valores predeterminados se usan para los otros dos.
    * Si especifica dos nmeros enteros dentro del parntesis con rango(), se usan para "start" y "stop", y el valor predeterminado se usa para "step".
    * puede especificar los tres nmeros enteros para "start", "stop" y "step".


```python
# range ciao
for i in range(0,3):
    print('ciao')
    
# equivalente
for i in range(3):
    print('ciao')
```

# List Comprehensions

* En Python, puedes crear listas de forma muy rpida y concisa mediante list comprehensions
* Las listas por comprensin nos permiten crear una lista usando un bucle for en un solo paso.
* se contiene en [ ],se define la accion que se tiene que ejecutar seguida da un for clasico 

```python
# con un ciclo for se crea una nueva lista con los nombres en mayusculas
cities = ['new york city', 'mountain view', 'chicago', 'los angeles']
mayusculas = []
for ciudad in cities:
    mayusculas.append(ciudad.title())
print(mayusculas)
```

```python
# list comprehension
mayusculas2 =[ciudad.title() for ciudad in cities]
mayusculas2
```

# Conditionals in List Comprehensions
* se puede agregar condicionales a listas por comprensin (listcomps). 
* su ubicacion es despus del iterable **for ** usando la palabra clave **if** para verificar una condicin en cada iteracin.
* en el caso de usar mas condicionales **elif** **else** se ponen al inicio despues de la condicion

```python
# en caso de sentencias if se ubica despues del ciclo for
enteros  = [x**2 for x in range(9) if x%2 == 0]
enteros
```

```python
# en caso de sentencias if y else se ponen des pues del la 
# funcion que se quiere repetir con el ciclo for
[x if x%2==0 else x+10 for x in range(1,15)]
```

```python
# Utilice una lista por comprensin para crear una nueva lista de nombres 
#que contenga solo los nombres en minsculas.
names = ["Rick Sanchez", "Morty Smith", "Summer Smith", "Jerry Smith", "Beth Smith"]

first_names = [name.lower() for name in names]
print(first_names)
```
